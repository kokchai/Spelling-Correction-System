```python
import requests
import unicodedata

import numpy as np

from bs4 import BeautifulSoup
from tqdm.notebook import tqdm
```

```python
N_PAGES = 10

links = list()
texts = list()

for i in tqdm(range(1, N_PAGES+1)):
    main_url = "https://blogs.esa.int/exploration/archive/page/"+str(i)
    
    main_page = requests.get(main_url)
    main_soup = BeautifulSoup(main_page.content, "html.parser")
    
    for p in main_soup.find_all("a", class_="c-button", align=""):
        links.append(p.get("href"))

for l in tqdm(links):
    page_url = l
    
    page_cont = requests.get(page_url)
    page_soup = BeautifulSoup(page_cont.content, "html.parser")
    
    for p in page_soup.find_all("p", class_="", align=""):
        texts.append(p.get_text())

for i, t in enumerate(texts):
    texts[i] = unicodedata.normalize("NFKD", t)
    texts[i] = t.replace("\n", " ")
    
texts = [t for t in texts if t != ""]
texts = [t for t in texts if t != " "]

texts = np.save("esablog.npy", np.array(texts))
```

```python
nasa = [("https://blogs.nasa.gov/spacestation/page/", 100),
        ("https://blogs.nasa.gov/mars2020/page/", 5),
        ("https://blogs.nasa.gov/parkersolarprobe/page/", 10),
        ("https://blogs.nasa.gov/artemis/page/", 10)]

texts = list()
for p in nasa:
    main_page = p[0]
    for j in tqdm(range(1, p[1]+1)):
        url = main_page+str(j)

        page = requests.get(url).content
        soup = BeautifulSoup(page, "html.parser")

        for p in soup.find_all("div", class_="entry-content", align=""):
            texts.append(p.get_text())

for i, t in enumerate(texts):
    texts[i] = unicodedata.normalize("NFKD", t)
    texts[i] = t.replace("\n", " ")

texts = [t for t in texts if t != ""]
texts = [t for t in texts if t != " "]

texts = np.save("nasablog.npy", np.array(texts))
```
